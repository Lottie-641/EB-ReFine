```python
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np
```

```python
file_path = 'v2.1_0.4_prefix/BPIC15_1.txt'
#file_path = 'helpdesk.txt' 
with open(file_path, 'r') as f:
    lines = f.readlines()[1:]  # Skip the header
```

```python

length_to_scores = defaultdict(lambda: {'no_freq': [], 'with_freq': []})

for line in lines:
    cols = line.strip().split(',')
    if len(cols) < 7:
        continue
    prefix = [x for x in cols[0].split() if x.isdigit()]
    prefix_len = len(prefix)
    try:
        dl_no_freq = float(cols[4])
        dl_with_freq = float(cols[5])
    except ValueError:
        continue
    length_to_scores[prefix_len]['no_freq'].append(dl_no_freq)
    length_to_scores[prefix_len]['with_freq'].append(dl_with_freq)
```

```python

sorted_lengths = sorted(length_to_scores.keys())
avg_no_freq = [np.mean(length_to_scores[l]['no_freq']) for l in sorted_lengths]
avg_with_freq = [np.mean(length_to_scores[l]['with_freq']) for l in sorted_lengths]
```

```python

# Step 6: Plotting with customization
plt.figure(figsize=(6, 2.7))
# plt.plot(sorted_lengths, avg_with_freq, color='#9B2227', marker='s', markersize=2, linewidth=2, label='ReFine(ours)')
# plt.plot(sorted_lengths, avg_no_freq, color='#073E7F', marker='o', markersize=2, linewidth=2, label='LUPIN')
plt.plot(sorted_lengths, avg_no_freq, color='#073E7F', marker='o', markersize=2.6, linewidth=2.6, label='LUPIN')
plt.plot(sorted_lengths, avg_with_freq, color='#9B2227', marker='s', markersize=2.6, linewidth=2.6, label='ReFine(ours)')
# Customize labels and title
#plt.xlabel('Prefix Length', fontsize=16)
plt.ylabel('Average DL', fontsize=18)
plt.title('BPIC15_1', fontsize=22)

# Customize ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Customize legend
#plt.legend( fontsize=16)

# Dashed grid
plt.grid(True, linestyle='--', linewidth=0.8)

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np

# Step 1: Load data
file_path = '../sepsis.txt'  # BPIC15_1.txt, bpic2017_o.txt, bpic2020.txt, helpdesk.txt,mip.txt, sepsis.txt
with open(file_path, 'r') as f:
    lines = f.readlines()[1:]  # Skip header

# Step 2: Group scores by truth length
truthlen_to_scores = defaultdict(lambda: {'no_freq': [], 'with_freq': []})

for line in lines:
    cols = line.strip().split(',')
    if len(cols) < 7:
        continue

    truth = [x for x in cols[3].split() if x.isdigit()]
    truth_len = len(truth)
    try:
        dl_nofreq = float(cols[4])
        dl_withfreq = float(cols[5])
    except ValueError:
        continue

    truthlen_to_scores[truth_len]['no_freq'].append(dl_nofreq)
    truthlen_to_scores[truth_len]['with_freq'].append(dl_withfreq)

# Step 3: Compute max truth length
max_truth_len = max(truthlen_to_scores.keys())

# Step 4: Map to prefix length and compute averages
prefix_to_scores = {}
for tlen, scores in truthlen_to_scores.items():
    prefix_len = max_truth_len - tlen
    prefix_to_scores[prefix_len] = {
        'no_freq': np.mean(scores['no_freq']),
        'with_freq': np.mean(scores['with_freq'])
    }

# Step 5: Sort and prepare for plot
sorted_prefixes = sorted(prefix_to_scores.keys())
avg_nofreq = [prefix_to_scores[p]['no_freq'] for p in sorted_prefixes]
avg_withfreq = [prefix_to_scores[p]['with_freq'] for p in sorted_prefixes]

# Step 6: Plotting with customization
plt.figure(figsize=(6, 2.5))

sorted_prefixes_trimmed = sorted_prefixes[:-1]
avg_nofreq_trimmed = avg_nofreq[:-1]
avg_withfreq_trimmed = avg_withfreq[:-1]

plt.plot(sorted_prefixes, avg_nofreq, color='#073E7F', marker='o', markersize=2, linewidth=2, label='LUPIN')
plt.plot(sorted_prefixes, avg_withfreq, color='#E0271E', marker='s', markersize=2, linewidth=2, label='ReFine(ours)')

# Customize labels and title
plt.xlabel('Prefix Length', fontsize=14)
plt.ylabel('Average DL', fontsize=14)
plt.title('Helpdesk', fontsize=16)

# Customize ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Customize legend
plt.legend(fontsize=12)

# Dashed grid
plt.grid(True, linestyle='--', linewidth=0.8)

# Tight layout
plt.tight_layout()

# Show plot
plt.show()

```
